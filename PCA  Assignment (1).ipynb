```python
#Converting into numpy array
data_array=data1.values
data_array
```

```python
#Normalizing the numerical data
data_norm=scale(data_array)
data_norm
```

```python
#Applying PCA
pca=PCA(n_components=13)
pca_values=pca.fit_transform(data_norm)
pca_values
```

```python
#Weights
pca.components_
```

```python
#The amount of variance that each PCA explains is 
var=pca.explained_variance_ratio_
var
```

```python
#Cumulative variance 
var1=np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
# Variance plot for PCA components obtained  Scree plot
plt.plot(var1,color="blue")
```

```python
PC = range(1, pca.n_components_+1)
plt.bar(PC, var, color='red')
plt.xlabel('Principal Components')
plt.ylabel('Variance %')
plt.xticks(PC)
plt.show()
```

## Clustering Algorithms

### 1.Hierarchical Clustering

```python
#Creating Dendrogram
plt.figure(figsize=(10,8))
dendrogram = sch.dendrogram(sch.linkage(data_norm,method='complete'))
```

```python
#Creating Clusters
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'complete')
hc
```

```python
y_hc = hc.fit_predict(data_norm)
y_hc
```

```python
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
data2=data.copy()
data2
```

```python
data2['Clusters']=Clusters
data2
```

```python
data2['Clusters'].value_counts()
```

### 2.KMeans Clustering

```python
#Finding optimum number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 3 clusters from above plot
model=KMeans(n_clusters=3)
model.fit(data_norm)
```

```python
model.labels_
```

```python
data3=data.copy()
data3
```

```python
#Assign clusters to the data set
data3['clusterid_new'] = model.labels_
data3
```

```python
data3['clusterid_new'].value_counts()
```

```python
data3.groupby(data3.clusterid_new).mean()
```
